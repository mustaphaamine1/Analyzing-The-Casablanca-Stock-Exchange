### Step 2:

```
# Importing the required modules
import os
import sys
import pandas as pd
from bs4 import BeautifulSoup

path = '/work/Dataset/DIAC_SALAF.aspx'

# empty list
data = []

# for getting the header from
# the HTML file
list_header = []
soup = BeautifulSoup(open(path),'html.parser')
header = soup.find_all("table")[0].find("tr")

for items in header:
	try:
		list_header.append(items.get_text())
	except:
		continue

# for getting the data
HTML_data = soup.find_all("table")[0].find_all("tr")[1:]

for element in HTML_data:
	sub_data = []
	for sub_element in element:
		try:
			sub_data.append(sub_element.get_text())
		except:
			continue
	data.append(sub_data)

# Storing the data into Pandas
# DataFrame
data = pd.DataFrame(data = data, columns = list_header)
```

### Step3:

Now that we have the dataframe we need to rename the columns :

```
data.columns=['date','closing','adjusted','evolution','quantity','volume']
```

```
data
```

Now we need to clean the data by changing the data type of each  column and also remove '\n' and change 26,25 to 26.25

```
data=data.apply(lambda x: x.str.replace(',','.'))
data=data.apply(lambda x: x.str.replace('\n',''))
data['closing']=data['closing'].astype('float')
data['adjusted']=data['adjusted'].astype('float')
data['evolution']=data['evolution'].astype('float')
data['quantity']=data['quantity'].astype('float')
data['volume']=data['volume'].astype('float')
data['date']= pd.to_datetime(data['date'])
```

```
data
```

Now the dataset is clean, we need to sort it by date and reset the index

```
data=data.sort_values('date')
data=data.reset_index()
data=data.drop('index',axis=1)
```
