### Step2:

Now let's create a function that plots the same graph like in the website using the plotly library:

```
import matplotlib.pyplot as plt
import numpy as np
class Stock:
    def __init__(self, name):
        self.name = name
        self.data = create_dataframe(name+'.aspx')

    def visualize(self):
        fig, ax1 = plt.subplots(figsize=(10, 5))
        ax2 = ax1.twinx()
        ax1.plot(self.data['date'], self.data['closing'], label='Close', color='b')
        ax2.plot(self.data['date'], self.data['volume'], label='Volume', color='r')
        ax1.set_xlabel('Date')
        ax1.set_ylabel('Close Price')
        ax2.set_ylabel('Volume')
        plt.title(f'{self.name} Stock Price and Volume')
        ax1.legend()
        ax2.legend()
        ax1.yaxis.set_ticks(range(int(0), int(2500), 500))
        #ax2.yaxis.set_ticks(range(int(1000), int(20000000), 50000))
        plt.show()
```

```
s1 = Stock("EQDOM")
s1.visualize()
```
